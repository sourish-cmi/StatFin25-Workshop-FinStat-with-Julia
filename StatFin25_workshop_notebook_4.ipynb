# **Capital Asset Pricing Model (CAPM)**

- To measure the performance of a particular security or fund relative to a benchmark portfolio, **Security Characteristic Lines (SCL)** are widely used.

- The SCL equation is

$$
\mathbb{E}(R_i) - R_F = \alpha_i + \beta_i \big( \mathbb{E}(R_M) - R_F \big),
$$

where $\alpha_i$ represents the **active (abnormal) return** of asset $i$.

---

- Suppose there are $N$ risky assets with returns $R_1, R_2, \ldots, R_N$ and portfolio weights $\omega_1, \omega_2, \ldots, \omega_N$.

- The return of the portfolio is

$$
R_P = \omega_1 R_1 + \omega_2 R_2 + \cdots + \omega_N R_N.
$$

---

- Let $R_M$ denote the return of the market index. According to the **Security Market Line (SML)**,

$$
\mathbb{E}(R_i) = R_F + \alpha_i + \beta_i \big( \mathbb{E}(R_M) - R_F \big).
$$

- Therefore,

$$
\mathbb{E}(R_P)
= R_F
+ \left( \sum_{i=1}^N \omega_i \alpha_i \right)
+ \left( \sum_{i=1}^N \omega_i \beta_i \right)
\big( \mathbb{E}(R_M) - R_F \big).
$$

That is,

$$
\mathbb{E}(R_P)
= R_F + \alpha_P + \beta_P \big( \mathbb{E}(R_M) - R_F \big),
$$

where

$$
\alpha_P = \sum_{i=1}^N \omega_i \alpha_i,
\qquad
\beta_P = \sum_{i=1}^N \omega_i \beta_i.
$$

---

- If CAPM holds, then the **expected price** of the portfolio satisfies

$$
\mathbb{E}(P_t \mid P_{t-1})
= P_{t-1} \exp \left\{
R_F + \alpha_P + \beta_P \mathbb{E}(R_M - R_F)
\right\},
$$

where the portfolio log-return is defined as

$$
R_P = \log(P_t) - \log(P_{t-1}).
$$


# **Factor Model**

- Factor models generalise the CAPM by allowing **more factors** than just market risk, in addition to the unexplained (idiosyncratic) risk of each asset.

- The factor model is defined as

$$
R_i - R_F = \beta_{0i} + \beta_{1i} x_1 + \cdots + \beta_{pi} x_p + \varepsilon_i,
$$

where $x_1, x_2, \ldots, x_p$ are the $p$ factors.


- Typically, there are **two major objectives** when using factor models:

  - If the objective is **prediction of returns**, any factor with good predictive power can be used.  
    A model with strong predictive power increases the chance of higher investment returns.

  - For example, experts in behavioural finance and data mining often measure **market sentiment** and include it as a factor due to its predictive ability.

  - However, if the objective is **understanding the relationship** between returns and factors, one should seek a **theoretical justification** for including each factor.


## **Factor Models in Matrix Notation**

- The factor model can be written in matrix form as

$$
\mathbf{r}_t = \mathbf{B} \mathbf{x}_t + \boldsymbol{\varepsilon}_t.
$$


where

$$
\mathbf{r}_t
=
\begin{bmatrix}
r_{t1} \\
\vdots \\
r_{tN}
\end{bmatrix}
=
\begin{bmatrix}
R_{t1} - R_F \\
\vdots \\
R_{tN} - R_F
\end{bmatrix},
$$

is the vector of **excess returns over the risk-free rate** at time $t$ for $N$ assets.


$$
\mathbf{B}
=
\begin{bmatrix}
\beta_{11} & \cdots & \beta_{1p} \\
\vdots & \ddots & \vdots \\
\beta_{N1} & \cdots & \beta_{Np}
\end{bmatrix}_{N \times p},
$$

is the **factor loading (sensitivity) matrix**.


$$
\mathbf{x}_t
=
\begin{bmatrix}
x_{t1} \\
\vdots \\
x_{tp}
\end{bmatrix},
$$

is the vector of **factor realisations** at time $t$.


- The error term $\boldsymbol{\varepsilon}_t$ captures **idiosyncratic (asset-specific) risk**.

## **Portfolio Return under a Factor Model**

- The portfolio return on day $t$ is

$$
r_{Pt} = \boldsymbol{\omega}^\top \mathbf{r}_t
       = \sum_{i=1}^{N} \omega_i r_{ti},
$$

where $\boldsymbol{\omega}^\top = (\omega_1, \ldots, \omega_N)$ is the **portfolio weight vector**,  
$\sum_{i=1}^{N} \omega_i = 1$, and $\omega_i$ is the weight of the $i$-th asset.

---

- Using the factor model, this can be expressed as

$$
r_{Pt} = \boldsymbol{\omega}^\top \mathbf{B} \mathbf{x}_t
       + \boldsymbol{\omega}^\top \boldsymbol{\varepsilon}_t.
$$

## **Expected Portfolio Return**

- The expected portfolio return is

$$
\mathbb{E}(r_P) = \boldsymbol{\omega}^\top \mathbf{B} \mathbf{x}
                = \boldsymbol{\beta}^\top \mathbf{x},
$$

where

$$
\boldsymbol{\beta}^\top = \boldsymbol{\omega}^\top \mathbf{B}
$$

is the **portfolio’s sensitivity (beta)** to the risk factors.


## **Portfolio Variance**

- The portfolio variance is given by

$$
\mathrm{Var}(r_P)
= \boldsymbol{\omega}^\top \mathbf{B} \boldsymbol{\Sigma}_x \mathbf{B}^\top \boldsymbol{\omega}
  + \boldsymbol{\omega}^\top \boldsymbol{\Sigma}_\varepsilon \boldsymbol{\omega},
$$

which can be written compactly as

$$
\mathrm{Var}(r_P)
= \boldsymbol{\omega}^\top
  \left( \mathbf{B} \boldsymbol{\Sigma}_x \mathbf{B}^\top
       + \boldsymbol{\Sigma}_\varepsilon \right)
  \boldsymbol{\omega}
= \boldsymbol{\omega}^\top \boldsymbol{\Sigma}_P \boldsymbol{\omega}.
$$

---

where:

- $\boldsymbol{\Sigma}_x$ is the **covariance matrix of the risk factors**,  
- $\boldsymbol{\Sigma}_\varepsilon$ is a **diagonal matrix** of idiosyncratic (unexplained) variances,  
- $\mathbf{B} \boldsymbol{\Sigma}_x \mathbf{B}^\top$ represents **systematic risk**,  
- $\boldsymbol{\Sigma}_P$ is the **portfolio covariance matrix**.


## **Benefit of Diversification**

- The unexplained (idiosyncratic) component of portfolio variance is

$$
\boldsymbol{\omega}^\top \boldsymbol{\Sigma}_\varepsilon \boldsymbol{\omega}.
$$

- If the portfolio is **equally weighted**, i.e.,

$$
\omega_i = \frac{1}{N}, \quad \forall i,
$$

then

$$
\boldsymbol{\omega}^\top \boldsymbol{\Sigma}_\varepsilon \boldsymbol{\omega}
= \frac{1}{N^2} \sum_{i=1}^{N} \sigma_i^2
\le \frac{\sigma_M^2}{N},
$$

where

$$
\sigma_M^2 = \max(\sigma_1^2, \ldots, \sigma_N^2) < \infty.
$$

- Therefore,

$$
\boldsymbol{\omega}^\top \boldsymbol{\Sigma}_\varepsilon \boldsymbol{\omega}
\;\longrightarrow\; 0 \quad \text{as } N \to \infty.
$$



```julia
using YFinance
using DataFrames
using Statistics
using LinearAlgebra
using Dates
using GLM
```

```julia
start_date = Date(2015,1,1)
end_date   = today()

ticker = "RELIANCE.NS"

mkt = get_prices("^NSEI", start_date, end_date)     # Market (NIFTY)
stk = get_prices(ticker, start_date, end_date)     # Stock
```

```julia

# Make separate dataframes (each one internally consistent)
df_mkt = DataFrame( Date = mkt["timestamp"]
                   ,Market = mkt["adjclose"])

df_stk = DataFrame( Date  = stk["timestamp"]
                   ,Stock = stk["adjclose"])

# Align on common trading dates
df = innerjoin(df_stk, df_mkt, on = :Date)

# Optional: sort by date (good practice)
sort!(df, :Date)

# Drop missings / non-finite values if any
dropmissing!(df)
filter!(row -> isfinite(row.Stock) && isfinite(row.Market), df)
last(df,7)
```

```julia
# log returns
df.R_stock = [missing; diff(log.(df.Stock))]
df.R_mkt   = [missing; diff(log.(df.Market))]

dropmissing!(df, [:R_stock, :R_mkt])

# risk-free (6.5% annual -> daily)
rf_daily = 0.065 / 252

df.excess_stock = df.R_stock .- rf_daily
df.excess_mkt   = df.R_mkt   .- rf_daily;
```

```julia
capm = lm(@formula(excess_stock ~ excess_mkt), df)
println(capm)

α = coef(capm)[1]
β = coef(capm)[2]

println("Alpha (annualised): ", α * 252)
println("Beta: ", β)
```

```julia
using Plots, Statistics

x = df.excess_mkt
y = df.excess_stock

# Scatter plot
scatter(
    x, y,
    xlabel = "NIFTY Excess Return",
    ylabel = "Reliance Excess Return",
    label = "Daily observations",
    alpha = 0.4,
    legend = :topleft
)
```

```julia
α̂ = coef(capm)[1]
β̂ = coef(capm)[2]

xgrid = range(minimum(x), maximum(x), length=200)
yfit  = α̂ .+ β̂ .* xgrid

plot!(
    xgrid, yfit,
    lw = 3,
    color = :blue,
    label = "CAPM fit"
)

plot!(grid = :on)
```

```julia
# Number of observations
n = nobs(capm)

# Number of regressors (excluding intercept)
k = length(coef(capm)) - 1

# Ordinary R-squared
R2 = r2(capm)   # this is the usual OLS R²

# Adjusted R-squared
adjR2 = 1 - (1 - R2) * (n - 1) / (n - k - 1)

println("R²        = ", R2)
println("Adj. R²   = ", adjR2)
```

```julia
resid = residuals(capm)

rse = sqrt(sum(resid.^2) / (n - k - 1))

println("Residual Std Error (RSE) = ", rse)
```
